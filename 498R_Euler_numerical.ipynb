```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("/Users/maryfagerburg/Downloads/Research/euler_output.csv")
```

```python
# analytical solution

def analytical(t):
    eq = np.exp(-3*t)
    return eq
```

```python
data["Analytical"] = analytical(data["t"])
data["Error"] = np.abs(data["Euler_x"] - data["Analytical"])
```

```python
step_size_1 = data[data["Step_size"] == 0.1]  #for step size 0.1
step_size_2 = data[data["Step_size"] == 0.15] #ect..
step_size_3 = data[data["Step_size"] == 0.2]
step_size_4 = data[data["Step_size"] == 0.25]
step_size_5 = data[data["Step_size"] == 0.3]
step_size_6 = data[data["Step_size"] == 0.35]
step_size_7 = data[data["Step_size"] == 0.4]
```
